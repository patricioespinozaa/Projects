# SONAR: Rock vs Mines Prediction Model using Linear Regression

Steps:
* Sonar Data
* Pre processing
* Train Test split
* Logistic Regression Model: It's useful since we will be working with binary classification

Dataset: https://www.kaggle.com/datasets/mayurdalvi/sonar-mine-dataset

## Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and Data Processing

```python
# load dataset
sonar_data = pd.read_csv('dataset/sonar_data.csv', header=None)
```

```python
# show first 5 rows of dataset
sonar_data.head()
```

```python
# number of rows and columns
print("rows and columns: ", sonar_data.shape)
# statistical measures of the data
sonar_data.describe()
```

M represents Mine and R represents Rock

```python
# Number of Mines and Rocks in the dataset
sonar_data[60].value_counts()
```

```python
# mean
sonar_data.groupby(60).mean()
```

```python
# separate data and labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]

print(X)
print(Y)
```

## Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

## Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

### Model Evaluation

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data : ', training_data_accuracy)

# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data : ', test_data_accuracy)
```
